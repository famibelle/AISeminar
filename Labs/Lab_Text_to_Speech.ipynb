```python
from transformers import pipeline

narrator = pipeline(
    "text-to-speech",
    model="kakao-enterprise/vits-ljs"
)
```

```python
text = """
The campaign has three complementary strands, each addressing the specific needs of its audience. Emma speaks to individuals, Marc to the self-employed, while the third section is aimed at employers concerned about the future of their teams. The tagline ‘Are you?’ naturally opens up the dialogue, inviting reflection without making people feel guilty.
Faced with a complex issue, LALUX is able to offer practical solutions. The message is clear: preparing for retirement means building for the future while benefiting from immediate tax savings. This two-pronged approach responds to current challenges, at a time of growing uncertainty surrounding the pensions model.
"""
```

```python
narrated_text = narrator(text)
```

```python
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
```
